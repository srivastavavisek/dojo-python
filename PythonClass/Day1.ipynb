***\*args***: allows function to accept arbitary number of positional arguments. Inside the function, args is a tuple

***\*kwargs***: arbitary keyword arguments. Inside the function, kwargs is a dictionary

````python
def calculate(calc, *args):
    match calc:
        case "add":
            result = sum(args)
        case "subtract":
            result = args[0]
            for i in args[1:]:
                result -= i
        case "multiply":
            result = 1
            for i in args:
                result *= i
        case _:
            raise ValueError(f"Invalid Operation: {calc}")
    return result

print(f"Adddition: {calculate("add", 1, 2, 3)}")
print(f"Subtraction: {calculate("subtract", 1, 2, 3)}")
print(f"multiply: {calculate("multiply", 1, 2, 3)}")
````
