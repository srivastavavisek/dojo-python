```python
# factorial-loop
i = 1
n = 5
while n >= 1:
    i = i*n
    n = n-1

print(i)
```

```python
# factorial-recursive
def fact(num):
    if num==1:
        return 1
    else:
        return num*fact(num-1)

print(fact(5))
```

```python
# Pattern
for i in range(1, 6):
    for j in range(1, i+1):
        print(i, end=" ")
    print()
```

```python
# Wap to find sum of digits of a given number
num = 12345
sum = 0
while num != 0:
    sum += num % 10
    num = num // 10

sum
```

```python
# Sum of first n numbers
def sumUpTo(n):
    sum = 0
    for i in range(1, n+1):
        sum += i
    return sum

print(sumUpTo(1000))
```

### **\*args and \*kwargs**

```python
def func(*args):
    print(args)
    print(type(args))

func(1, 2, 3)
```

```python
def func(**kwargs):
    print(kwargs)
    print(type(kwargs))

func(a=10, b=20, c=30)
```

```python
def func(m, n, *args, **kwargs):
    print(m, n)
    print(args)
    print(kwargs)

func(1, 2, 3, 4, 5, a=10, b=20, c=30)
```

```python
def avg(*args):
    sum = 0
    for i in args:
        sum += i
    return sum/len(args)

print(avg(7, 1, 2, 3, 4, 5, 6))
```

```python
def func(**kwargs):
    sum = 0
    for key in kwargs.keys():
        sum += kwargs[key]
    return sum

func(a=10, b=20, c=30)
```

```python
# calculator
def calculate(a, b):
    sum = a+b
    diff = a-b
    prod = a*b
    quo = a/b
    return sum, diff, prod, quo

result = calculate(10, 2)
sum, diff, prod, quo = result

print(result)
print(sum, diff, prod, quo)
print(type(result))
```

##### unpacking

```python
def get_name(first, middle, last):
    print(f"first name = {first}")
    print("middle name = ", middle)
    print("last name = ", last)

aList = ["Abhishek", "Kumar", "Kayastha"]
get_name(*aList)
```

```python
aDict = {"first": "Abhishek", "middle": "Kumar", "last": "Kayastha"}
get_name(**aDict)
```
