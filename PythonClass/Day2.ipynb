```python
class Distance:
    def __init__(self, feet, inches):
        self.feet = feet
        self.inches = inches
        self.total = self.feet * 12 + self.inches

    def __str__(self):
        return f'{self.total // 12} feet and {self.inches % 12} inches'
    
    def __add__(self, second):
        result = self.total + second.total
        return Distance(result // 12, result % 12)
    
    def __sub__(self, second):
        result = self.total - second.total
        return Distance(result // 12, result % 12) if (result > 0) else Distance(-result // 12, -result % 12)
```

```python
    
d1 = Distance(2, 2)
d2 = Distance(3, 5)
d3 = d1 + d2
d4 = d1 - d2
print(f'Sum: {d3}')
print(f'Difference: {d4}')
```
