```python
class Animal:
    def __init__(self, name, voice):
        self.name = name
        self.voice = voice

    def __str__(self):
        return self.name

    def __repr__(self):
        return f"Animal: {self.name}"

    def bark(self):
        print({self.voice})
```

```python
cat = Animal("billie", "meow")

print(cat)
cat.bark()
cat
```

***Python class Distance to represent distances in feet and inches, also overload any two operators***

```python
class Distance:
    def __init__(self, feet=0, inches=0):
        self.total = feet * 12 + inches
        self.feet = self.total // 12
        self.inches = self.total % 12

    def __str__(self):
        return f'{self.feet} feet and {self.inches} inches'
    
    def __add__(self, second):
        result = self.total + second.total
        return Distance(inches=result)
    
    def __sub__(self, second):
        result = self.total - second.total
        return Distance(inches=result) if (result > 0) else Distance(inches=-result)
```

```python
d1 = Distance(1, 2)
d2 = Distance(3, 5)
d3 = d1 + d2
d4 = d1 - d2
print(f'Sum: {d3}')
print(f'Difference: {d4}')
```

***More operator overloadings***

```python
def __eq__(self, second):
    return self.total == second.total

Distance.__eq__ = __eq__

dd1 = Distance(4, 5)
dd2 = Distance(3, 17)
print("Equal") if (dd1 == dd2) else print ("Not Equal")
```
